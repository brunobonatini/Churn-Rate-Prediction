A retenção de clientes é uma das principais preocupações para empresas de telecomunicações, uma vez que a perda de clientes, conhecida como **Churn**, pode ter um impacto significativo nos negócios. A previsão de churn tornou-se uma ferramenta valiosa para essas empresas, permitindo antecipar quais clientes estão mais propensos a cancelar seus serviços.

**Churn rate**, ou simplesmente **Churn**, representa a taxa de evasão da sua base de clientes. Em serviços como, Vivo, Claro, Spotify ou Netflix, ela representaria a taxa de cancelamento de assinaturas.

Monitorar e entender o **Churn** é de extrema importância e sua análise ao longo do tempo pode mostrar que existe algum problema que deve ser atacado. Também pode ser usado para identificar potenciais de cancelamentos com um tempo de antecedência, e promover ações direcionadas para tentar reter tais clientes. Essa métrica deve receber atenção pelo fato de que o Custo de Aquisição de Cliente (CAC) é normalmente mais alto que o custo para mantê-los. Ou seja, um alto valor para o **Churn** é o que não desejamos.

## Objetivo do Projeto

Utilizar técnicas de Análise de dados e Machine Learning para identificar padrões e tendências que possam indicar a probabilidade de um cliente cancelar seu contrato, para que as empresas de telecomunicações consigam implementar estratégias proativas de retenção, como oferecer ofertas personalizadas, melhorar a qualidade do atendimento ao cliente, ajustar preços ou fornecer serviços adicionais que atendam às necessidades específicas do cliente.
